# Notebook: Clasificación Binaria con Regresión Logística

## 1.1 Fundamentos de la técnica (Logistic Regression)
La **Regresión Logística** es un método estadístico que se utiliza principalmente para **clasificación binaria** (predicción de dos clases, e.g., Sí/No). A pesar de su nombre, no es un algoritmo de regresión, sino de clasificación.

El modelo funciona estimando la probabilidad de que ocurra un evento binario. Utiliza la **función Sigmoide**  para transformar el resultado lineal en una probabilidad entre 0 y 1.

## 1.2 Modelo Matemático de la técnica
La Regresión Logística modela la probabilidad $P$ de pertenecer a la clase 1 (Clase Positiva).

1.  **Función Lineal (Z):** Calcula una combinación lineal de las variables de entrada ($X$) con los coeficientes ($\beta$):
    $$
    Z = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n
    $$

2.  **Función Sigmoide ($\sigma$):** Transforma $Z$ en la probabilidad $P$ de pertenecer a la clase 1:
    $$
    P(y=1|X) = \sigma(Z) = \frac{1}{1 + e^{-Z}}
    $$
    * **Decisión:** Si la probabilidad calculada ($P$) es **mayor a 0.5** (el umbral de decisión), se predice la clase 1.

## 1.3 Descripción de la librería, clases, funciones y su parámetros (Python)

### Librerías Base
* **`pandas` (pd):** Utilizada para la **carga, estructuración y manipulación** de datos tabulares (DataFrames). Es esencial para leer el archivo CSV.
* **`numpy` (np):** Proporciona soporte para **matrices y arrays** multidimensionales, la base para todas las operaciones matemáticas del modelo.
* **`matplotlib.pyplot` (plt) y `seaborn` (sns):** Librerías de **visualización** utilizadas para generar el Gráfico de Dispersión exploratorio y el *heatmap* de la Matriz de Confusión.

### Clases y Funciones de Scikit-learn (`sklearn`)
* **`LogisticRegression` (sklearn.linear_model):** La clase principal. Se entrena buscando los coeficientes que **maximizan la verosimilitud**. El parámetro `C` (inverso de la fuerza de regularización) es clave para prevenir el *overfitting*.
* **`StandardScaler` (sklearn.preprocessing):** **Esencial.** Utilizado para **normalizar los datos** (transformación Z-score). El escalado asegura que el algoritmo de optimización converja eficientemente, ya que el modelo Logístico es sensible a la escala de las características.
* **`train_test_split` (sklearn.model_selection):** Divide el *dataset* en conjuntos de entrenamiento y prueba para una **evaluación imparcial** del rendimiento.
* **Métricas:** `confusion_matrix`, `accuracy_score`, `classification_report` se utilizan para la evaluación final.

```python
# Importación de Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Cargar dataset de Pima Indians Diabetes (8 features, 1 target binario)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'insu', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(url, names=names)
```

## 1.4. Pipeline

### Preprocesamiento
Se realiza el análisis exploratorio y el Gráfico de Dispersión para comprender la estructura y el balance de clases del dataset Pima Diabetes.

```python
# 1. Análisis estadístico y estructura de los datos
print("Estructura de las primeras filas:")
display(df.head())
print("\nConteo de clases (0: No Diabetes, 1: Diabetes) - balance:")
print(df['class'].value_counts())

# 2. Gráfico de Dispersión (Scatter Plot)
# Se visualiza la relación entre las dos características más predictivas: Glucosa (plas) y Edad (age).
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="age",
    y="plas", 
    hue="class",
    palette={0: 'blue', 1: 'red'},
    s=50
)
plt.title("Relación entre Edad y Glucosa (Pima Diabetes)")
plt.xlabel("Edad")
plt.ylabel("Glucosa en Plasma")
plt.legend(title="Clase")
plt.grid(True)
plt.show()
```

### Feature Engineering
* **Selección de Variables:** Se utilizan las 8 características médicas como entradas ($X$) para predecir el diagnóstico ($y$).
* **Escalado:** Se aplica el **Escalado Estándar** (`StandardScaler`). Este escalado es crucial porque asegura que el rango de valores (e.g., Glucosa vs. Edad) no distorsione el cálculo de los coeficientes del modelo.

```python
# 1. Separar variables X y y
X = df.drop('class', axis=1).values 
y = df['class'].values              

# 2. División Train/Test (30% para prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=16 # El random state garantiza reproducibilidad
)

# 3. Escalado (fit en train, transform en test)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Datos escalados listos para el entrenamiento. Total de features:", X.shape[1])
```

### Model Training
Se instancia el clasificador `LogisticRegression` y se entrena con los datos estandarizados. El proceso de `fit` encuentra los 8 coeficientes ($\beta_1$ a $\beta_8$) y el intercepto ($\beta_0$) que optimizan la probabilidad de observar las clases reales.

```python
# Instanciar y entrenar el modelo
model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_scaled, y_train)

print("Modelo de Regresión Logística entrenado con éxito.")
print("\nParámetros del modelo ajustados (8 Coeficientes + Intercepto):")
print(f"Intercepto (β0): {model_lr.intercept_[0]:.4f}")
print(f"Coeficientes (β1-β8, escalados): {model_lr.coef_[0]}")
```

### Prediction
Se crea una función de Python para clasificar un nuevo patrón. Esta función debe ser robusta, aplicando el escalado aprendido en el entrenamiento al patrón de entrada antes de la predicción.

```python
# Función para clasificar un nuevo patrón
def predecir_diagnostico(preg, plas, pres, skin, insu, mass, pedi, age):
    """
    Recibe las 8 características del dataset Pima en unidades originales, las escala y predice el diagnóstico.
    """
    # 1. Crear un array 2D con el patrón de entrada
    features_list = [preg, plas, pres, skin, insu, mass, pedi, age]
    sample = np.array([features_list])
    
    # 2. Aplicar el escalado ENTRENADO (¡Paso fundamental!)
    sample_scaled = scaler.transform(sample)
    
    # 3. Predicción (devuelve 0 o 1)
    pred_idx = model_lr.predict(sample_scaled)[0]
    
    # 4. Obtener el diagnóstico y la probabilidad
    diagnostico = "Diabetes" if pred_idx == 1 else "No Diabetes"
    probabilidad = model_lr.predict_proba(sample_scaled)[0, pred_idx]
    
    print(f"\n--- Predicción para Nuevo Patrón ---")
    print(f"Características de entrada: {sample[0]}")
    print(f"Diagnóstico predicho: {diagnostico} ({pred_idx})")
    print(f"Probabilidad de la clase: {probabilidad:.4f}")
    
    return diagnostico

# Ejemplo de uso (valores de una paciente que, en la vida real, NO tuvo diabetes)
predecir_diagnostico(preg=1, plas=89, pres=66, skin=23, insu=94, mass=28.1, pedi=0.167, age=21)
```

### Model Evaluation
Se evalúa el modelo en el conjunto de prueba (`X_test_scaled`), con explicaciones detalladas de los fundamentos y algoritmos de las métricas.

**Métricas de Clasificación (Fundamentos y Algoritmos):**

1.  **Matriz de Confusión:**
    * **Fundamento:** Tabla que resume el desempeño comparando los valores reales vs. los predichos. Es la base para calcular todas las demás métricas. La **diagonal principal** muestra los aciertos (VP y VN).
    * **Contenido:** Muestra los aciertos en la **diagonal principal** (Verdaderos Positivos y Verdaderos Negativos) y los errores (Falsos Positivos y Falsos Negativos).

2.  **Accuracy (Exactitud):**
    * **Fundamento:** Mide el porcentaje de predicciones correctas sobre el total de muestras. Es una métrica de desempeño global, ideal para datasets balanceados.
    * **Algoritmo:** $$\text{Accuracy} = \frac{\text{Aciertos}}{\text{Total de Muestras}} = \frac{\text{VP} + \text{VN}}{\text{VP} + \text{VN} + \text{FP} + \text{FN}}$$

```python
# Generar predicciones para el conjunto de prueba
y_pred = model_lr.predict(X_test_scaled)

# 1. Matriz de Confusión
print("--- 1. Matriz de Confusión ---")
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
cm_df = pd.DataFrame(cm, index=['Real: No Diabetes (0)', 'Real: Diabetes (1)'], 
                     columns=['Predicho: No Diabetes (0)', 'Predicho: Diabetes (1)'])
display(cm_df)

# Visualización de la matriz
plt.figure(figsize=(5, 4))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="PuRd")
plt.title("Matriz de Confusión (Regresión Logística)")
plt.ylabel("Clase Real"); plt.xlabel("Clase Predicha")
plt.show()

# 2. Accuracy Score
acc = accuracy_score(y_test, y_pred)
print(f"\n--- 2. Accuracy (Exactitud Global) ---")
print(f"Accuracy: {acc:.4f}")

# 3. Reporte de Clasificación (Métricas detalladas para contexto)
print("\n--- Reporte Detallado de Clasificación (Precision, Recall, F1) ---")
print(classification_report(y_test, y_pred, target_names=['No Diabetes', 'Diabetes']))
```
